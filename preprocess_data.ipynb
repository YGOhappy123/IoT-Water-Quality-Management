# PTIT HCM IOT - Water Quality Management

## Data prepocessing

```python
import pandas as pd

raw_df = pd.read_csv('dataset/raw_dataset.csv')
raw_df.head()
```

```python
def round_value (value):
    if isinstance(value, (int, float)):
        return round(value, 2)
        
    return value 
```

```python
raw_df['temperature'] = raw_df['temperature'].apply(round_value)
raw_df['water_level'] = raw_df['water_level'].apply(round_value)
raw_df['tds_level'] = raw_df['tds_level'].apply(round_value)
raw_df['env_humidity'] = raw_df['env_humidity'].apply(round_value)

raw_df.head()
```

```python
raw_df.to_csv('dataset/processed_dataset.csv', index=False, encoding='utf-8-sig')
```
